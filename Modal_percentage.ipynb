```python
import pandas as pd
import numpy as np
from scipy import linalg
from scipy.sparse import rand
from scipy.optimize import lsq_linear
```

```python
df = pd.read_excel('Total.xlsx', index_col=None)
df.head()
```

```python
df_mean = df.groupby(['Comment', 'Type']).mean()
df_mean.head()
```

```python
index_column=df_mean.index.values.tolist()
Comment=list(set([index_column[i][0] for i in range(len(index_column))]))
Comment.sort()
Comment
```

```python
list1=[54.0992831,1.3532258,16.4948484,9.3330848,0.1490751,5.3123065,8.7824767,3.3455042,0.8224647,0.2870586,0.0206703]
starting =np.array(list1)
```

```python
modal_table = pd.DataFrame()
for i in range(len(Comment)):
    mineral_comp=df_mean.loc[Comment[i]].transpose()[:11].to_numpy()
    x = lsq_linear(mineral_comp, starting, bounds=(0,1), lsmr_tol='auto', verbose=1)
    modal_ratio=list(x.values())[0].tolist()
    mineral_list=df_mean.loc[Comment[i]].index.values.tolist()
    dic = { name:value for name, value in zip(mineral_list, modal_ratio) }
    df_dictionary = pd.DataFrame([dic])
    modal_table = pd.concat([modal_table, df_dictionary], ignore_index=True)
```

```python
modal_table.index=Comment
modal_table = modal_table[["MG", "plg", "ol", "py", "Spinel"]]
modal_table.to_excel('modal_table.xlsx')
modal_table
```
